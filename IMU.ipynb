```python
import numpy as np
import random

def generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg):
    num_samples = int(duration * sampling_rate)
    time = np.linspace(0, duration, num_samples)

    # Generate simulated accelerations and angular velocities for both femurs
    acc_x_left = np.zeros(num_samples)
    acc_y_left = np.zeros(num_samples)
    acc_z_left = np.zeros(num_samples)
    gyro_x_left = np.zeros(num_samples)
    gyro_y_left = np.zeros(num_samples)
    gyro_z_left = np.zeros(num_samples)

    acc_x_right = np.zeros(num_samples)
    acc_y_right = np.zeros(num_samples)
    acc_z_right = np.zeros(num_samples)
    gyro_x_right = np.zeros(num_samples)
    gyro_y_right = np.zeros(num_samples)
    gyro_z_right = np.zeros(num_samples)

    # Simulate walking motion for both legs
    for i, t in enumerate(time):
        if disabled_leg == 'left':
            disabled_params = {
                'step_length': random.uniform(0.52, 0.57),
                'frequency': random.uniform(1.07, 1.09),
                'lateral_amplitude': random.uniform(0.19, 0.2),
                'forward_amplitude': random.uniform(0.09, 0.1),
                'angular_velocity_amplitude': random.uniform(22.0, 23.5),
                'acc_z_offset': 0.8,
                'gyro_y_offset': 6,
                'gyro_z_offset': 8
            }
            non_disabled_params = {
                'step_length': random.uniform(0.73, 0.77),
                'frequency': random.uniform(1.11, 1.13),
                'lateral_amplitude': random.uniform(0.2, 0.22),
                'forward_amplitude': random.uniform(0.12, 0.15),
                'angular_velocity_amplitude': random.uniform(30.3, 31.8),
                'acc_z_offset': 1.0,
                'gyro_y_offset': 10,
                'gyro_z_offset': 5
            }

            if t <= duration / 2:
                forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * t)
                lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)
                angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

                acc_x_left[i] = forward_motion_disabled
                acc_y_left[i] = lateral_motion_disabled

                gyro_x_left[i] = angular_velocity_disabled
                gyro_y_left[i] = disabled_params['gyro_y_offset'] * np.cos(2 * np.pi * disabled_params['frequency'] * t)
                gyro_z_left[i] = disabled_params['gyro_z_offset'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

                forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * t)
                lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
                angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)

                acc_x_right[i] = forward_motion_non_disabled
                acc_y_right[i] = lateral_motion_non_disabled

                gyro_x_right[i] = angular_velocity_non_disabled
                gyro_y_right[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
                gyro_z_right[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * t)

            else:
                forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * (duration - t))
                lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * (duration - t))
                angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * (duration - t))

                acc_x_left[i] = forward_motion_disabled
                acc_y_left[i] = lateral_motion_disabled

                gyro_x_left[i] = angular_velocity_disabled
                gyro_y_left[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                gyro_z_left[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

                forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * (duration - t))
                lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

                acc_x_right[i] = forward_motion_non_disabled
                acc_y_right[i] = lateral_motion_non_disabled

                gyro_x_right[i] = angular_velocity_non_disabled
                gyro_y_right[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                gyro_z_right[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

            acc_z_left[i] = disabled_params['acc_z_offset'] - 9.81
            if t % (1 / disabled_params['frequency']) < 0.5 / disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_left[i] += 0.5 * disabled_params['frequency'] * disabled_params['step_length'] * (1 - np.cos(2 * np.pi * disabled_params['frequency'] * t))

            acc_z_right[i] = non_disabled_params['acc_z_offset'] - 9.81
            if t % (1 / non_disabled_params['frequency']) < 0.5 / non_disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_right[i] += 0.5 * non_disabled_params['frequency'] * non_disabled_params['step_length'] * (1 - np.cos(2 * np.pi * non_disabled_params['frequency'] * t))

        elif disabled_leg == 'right':
            disabled_params = {
                'step_length': random.uniform(0.52, 0.57),
                'frequency': random.uniform(1.07, 1.09),
                'lateral_amplitude': random.uniform(0.19, 0.2),
                'forward_amplitude': random.uniform(0.09, 0.1),
                'angular_velocity_amplitude': random.uniform(22.0, 23.5),
                'acc_z_offset': 0.8,
                'gyro_y_offset': 6,
                'gyro_z_offset': 8
            }
            non_disabled_params = {
                'step_length': random.uniform(0.73, 0.77),
                'frequency': random.uniform(1.11, 1.13),
                'lateral_amplitude': random.uniform(0.2, 0.22),
                'forward_amplitude': random.uniform(0.12, 0.15),
                'angular_velocity_amplitude': random.uniform(30.3, 31.8),
                'acc_z_offset': 1.0,
                'gyro_y_offset': 10,
                'gyro_z_offset': 5
            }

            if t <= duration / 2:
                forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * t)
                lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)
                angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

                acc_x_right[i] = forward_motion_disabled
                acc_y_right[i] = lateral_motion_disabled

                gyro_x_right[i] = angular_velocity_disabled
                gyro_y_right[i] = disabled_params['gyro_y_offset'] * np.cos(2 * np.pi * disabled_params['frequency'] * t)
                gyro_z_right[i] = disabled_params['gyro_z_offset'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

                forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * t)
                lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
                angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)

                acc_x_left[i] = forward_motion_non_disabled
                acc_y_left[i] = lateral_motion_non_disabled

                gyro_x_left[i] = angular_velocity_non_disabled
                gyro_y_left[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
                gyro_z_left[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * t)

            else:
                forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * (duration - t))
                lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * (duration - t))
                angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * (duration - t))

                acc_x_right[i] = forward_motion_disabled
                acc_y_right[i] = lateral_motion_disabled

                gyro_x_right[i] = angular_velocity_disabled
                gyro_y_right[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                gyro_z_right[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

                forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * (duration - t))
                lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

                acc_x_left[i] = forward_motion_non_disabled
                acc_y_left[i] = lateral_motion_non_disabled

                gyro_x_left[i] = angular_velocity_non_disabled
                gyro_y_left[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * (duration - t))
                gyro_z_left[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * (duration - t))

            acc_z_right[i] = disabled_params['acc_z_offset'] - 9.81
            if t % (1 / disabled_params['frequency']) < 0.5 / disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_right[i] += 0.5 * disabled_params['frequency'] * disabled_params['step_length'] * (1 - np.cos(2 * np.pi * disabled_params['frequency'] * t))

            acc_z_left[i] = non_disabled_params['acc_z_offset'] - 9.81
            if t % (1 / non_disabled_params['frequency']) < 0.5 / non_disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_left[i] += 0.5 * non_disabled_params['frequency'] * non_disabled_params['step_length'] * (1 - np.cos(2 * np.pi * non_disabled_params['frequency'] * t))

    return time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
           acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right

# Parameters
duration = 10  # seconds
sampling_rate = 25  # Hz

# Generate IMU data for a disabled person with the disability in the left leg
time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right = \
    generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg='left')

# Alternatively, generate IMU data for a disabled person with the disability in the right leg
# time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
# acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right = \
#     generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg='right')

# Save data to CSV file
output_file = "imu_data_disabled_person.csv"
all_data = np.vstack((time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left,
                      acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right)).T
np.savetxt(output_file, all_data, delimiter=',',
           header='Time (s),Acc_X_Left (m/s^2),Acc_Y_Left (m/s^2),Acc_Z_Left (m/s^2),'
                  'Gyro_X_Left (deg/s),Gyro_Y_Left (deg/s),Gyro_Z_Left (deg/s),'
                  'Acc_X_Right (m/s^2),Acc_Y_Right (m/s^2),Acc_Z_Right (m/s^2),'
                  'Gyro_X_Right (deg/s),Gyro_Y_Right (deg/s),Gyro_Z_Right (deg/s)',
           fmt='%.5f', comments='')

print("IMU data for the disabled person saved to:", output_file)
```

```python
import numpy as np
import random

def generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg):
    num_samples = int(duration * sampling_rate)
    time = np.linspace(0, duration, num_samples)

    # Generate simulated accelerations and angular velocities for both femurs
    acc_x_left = np.zeros(num_samples)
    acc_y_left = np.zeros(num_samples)
    acc_z_left = np.zeros(num_samples)
    gyro_x_left = np.zeros(num_samples)
    gyro_y_left = np.zeros(num_samples)
    gyro_z_left = np.zeros(num_samples)

    acc_x_right = np.zeros(num_samples)
    acc_y_right = np.zeros(num_samples)
    acc_z_right = np.zeros(num_samples)
    gyro_x_right = np.zeros(num_samples)
    gyro_y_right = np.zeros(num_samples)
    gyro_z_right = np.zeros(num_samples)

    # Simulate walking motion for both legs
    for i, t in enumerate(time):
        if disabled_leg == 'left':
            disabled_params = {
                'step_length': random.uniform(0.52, 0.57),
                'frequency': random.uniform(1.07, 1.09),
                'lateral_amplitude': random.uniform(0.19, 0.2),
                'forward_amplitude': random.uniform(0.09, 0.1),
                'angular_velocity_amplitude': random.uniform(22.0, 23.5),
                'acc_z_offset': 0.8,
                'gyro_y_offset': 6,
                'gyro_z_offset': 8
            }
            non_disabled_params = {
                'step_length': random.uniform(0.73, 0.77),
                'frequency': random.uniform(1.11, 1.13),
                'lateral_amplitude': random.uniform(0.2, 0.22),
                'forward_amplitude': random.uniform(0.12, 0.15),
                'angular_velocity_amplitude': random.uniform(30.3, 31.8),
                'acc_z_offset': 1.0,
                'gyro_y_offset': 10,
                'gyro_z_offset': 5
            }

            forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * t)
            lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)
            angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

            acc_x_left[i] = forward_motion_disabled
            acc_y_left[i] = lateral_motion_disabled
            acc_z_left[i] = disabled_params['acc_z_offset'] - 9.81  # Gravity
            if t % (1 / disabled_params['frequency']) < 0.5 / disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_left[i] += 0.5 * disabled_params['frequency'] * disabled_params['step_length'] * (1 - np.cos(2 * np.pi * disabled_params['frequency'] * t))

            gyro_x_left[i] = angular_velocity_disabled
            gyro_y_left[i] = disabled_params['gyro_y_offset'] * np.cos(2 * np.pi * disabled_params['frequency'] * t)
            gyro_z_left[i] = disabled_params['gyro_z_offset'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

            forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * t)
            lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
            angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)

            acc_x_right[i] = forward_motion_non_disabled
            acc_y_right[i] = lateral_motion_non_disabled
            acc_z_right[i] = non_disabled_params['acc_z_offset'] - 9.81
            if t % (1 / non_disabled_params['frequency']) < 0.5 / non_disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_right[i] += 0.5 * non_disabled_params['frequency'] * non_disabled_params['step_length'] * (1 - np.cos(2 * np.pi * non_disabled_params['frequency'] * t))

            gyro_x_right[i] = angular_velocity_non_disabled
            gyro_y_right[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
            gyro_z_right[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * t)

        elif disabled_leg == 'right':
            disabled_params = {
                'step_length': random.uniform(0.52, 0.57),
                'frequency': random.uniform(1.07, 1.09),
                'lateral_amplitude': random.uniform(0.19, 0.2),
                'forward_amplitude': random.uniform(0.09, 0.1),
                'angular_velocity_amplitude': random.uniform(22.0, 23.5),
                'acc_z_offset': 0.8,
                'gyro_y_offset': 6,
                'gyro_z_offset': 8
            }
            non_disabled_params = {
                'step_length': random.uniform(0.73, 0.77),
                'frequency': random.uniform(1.11, 1.13),
                'lateral_amplitude': random.uniform(0.2, 0.22),
                'forward_amplitude': random.uniform(0.12, 0.15),
                'angular_velocity_amplitude': random.uniform(30.3, 31.8),
                'acc_z_offset': 1.0,
                'gyro_y_offset': 10,
                'gyro_z_offset': 5
            }

            forward_motion_disabled = disabled_params['forward_amplitude'] * np.cos(np.pi * disabled_params['frequency'] * t)
            lateral_motion_disabled = disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)
            angular_velocity_disabled = disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

            acc_x_right[i] = forward_motion_disabled
            acc_y_right[i] = lateral_motion_disabled
            acc_z_right[i] = disabled_params['acc_z_offset'] - 9.81  # Gravity
            if t % (1 / disabled_params['frequency']) < 0.5 / disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_right[i] += 0.5 * disabled_params['frequency'] * disabled_params['step_length'] * (1 - np.cos(2 * np.pi * disabled_params['frequency'] * t))

            gyro_x_right[i] = angular_velocity_disabled
            gyro_y_right[i] = disabled_params['gyro_y_offset'] * np.cos(2 * np.pi * disabled_params['frequency'] * t)
            gyro_z_right[i] = disabled_params['gyro_z_offset'] * np.sin(2 * np.pi * disabled_params['frequency'] * t)

            forward_motion_non_disabled = non_disabled_params['forward_amplitude'] * np.cos(np.pi * non_disabled_params['frequency'] * t)
            lateral_motion_non_disabled = non_disabled_params['lateral_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
            angular_velocity_non_disabled = non_disabled_params['angular_velocity_amplitude'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)

            acc_x_left[i] = forward_motion_non_disabled
            acc_y_left[i] = lateral_motion_non_disabled
            acc_z_left[i] = non_disabled_params['acc_z_offset'] - 9.81
            if t % (1 / non_disabled_params['frequency']) < 0.5 / non_disabled_params['frequency']:
                # Smooth transition for forward acceleration during swing phase
                acc_z_left[i] += 0.5 * non_disabled_params['frequency'] * non_disabled_params['step_length'] * (1 - np.cos(2 * np.pi * non_disabled_params['frequency'] * t))

            gyro_x_left[i] = angular_velocity_non_disabled
            gyro_y_left[i] = non_disabled_params['gyro_y_offset'] * np.sin(2 * np.pi * non_disabled_params['frequency'] * t)
            gyro_z_left[i] = non_disabled_params['gyro_z_offset'] * np.cos(2 * np.pi * non_disabled_params['frequency'] * t)

    return time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
           acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right

# Parameters
duration = 10  # seconds
sampling_rate = 100  # Hz

# Generate IMU data for a disabled person with the disability in the left leg
time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right = \
    generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg='left')

# Alternatively, generate IMU data for a disabled person with the disability in the right leg
# time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left, \
# acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right = \
#     generate_imu_data_disabled_person(duration, sampling_rate, disabled_leg='right')

# Save data to CSV file
output_file = "imu_data_disabled_person.csv"
all_data = np.vstack((time, acc_x_left, acc_y_left, acc_z_left, gyro_x_left, gyro_y_left, gyro_z_left,
                      acc_x_right, acc_y_right, acc_z_right, gyro_x_right, gyro_y_right, gyro_z_right)).T
np.savetxt(output_file, all_data, delimiter=',',
           header='Time (s),Acc_X_Left (m/s^2),Acc_Y_Left (m/s^2),Acc_Z_Left (m/s^2),'
                  'Gyro_X_Left (deg/s),Gyro_Y_Left (deg/s),Gyro_Z_Left (deg/s),'
                  'Acc_X_Right (m/s^2),Acc_Y_Right (m/s^2),Acc_Z_Right (m/s^2),'
                  'Gyro_X_Right (deg/s),Gyro_Y_Right (deg/s),Gyro_Z_Right (deg/s)',
           fmt='%.5f', comments='')

print("IMU data for the disabled person saved to:", output_file)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to apply moving average filter to data
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Load IMU data from the CSV file
data = np.loadtxt(output_file, delimiter=',', skiprows=1)

# Extract time and gyro data for the left and right legs
time = data[:, 0]
gyro_x_left = data[:, 4]
gyro_y_left = data[:, 5]
gyro_z_left = data[:, 6]
gyro_x_right = data[:, 10]
gyro_y_right = data[:, 11]
gyro_z_right = data[:, 12]

# Integrate gyro data to get angle data (assuming constant sampling rate)
angle_x_left = np.cumsum(gyro_x_left) * (time[1] - time[0])
angle_y_left = np.cumsum(gyro_y_left) * (time[1] - time[0])
angle_z_left = np.cumsum(gyro_z_left) * (time[1] - time[0])
angle_x_right = np.cumsum(gyro_x_right) * (time[1] - time[0])
angle_y_right = np.cumsum(gyro_y_right) * (time[1] - time[0])
angle_z_right = np.cumsum(gyro_z_right) * (time[1] - time[0])

# Smoothing window size (adjust as needed)
window_size = 10

# Apply moving average filter to angle data
smooth_angle_x_left = moving_average(angle_x_left, window_size)
smooth_angle_y_left = moving_average(angle_y_left, window_size)
smooth_angle_z_left = moving_average(angle_z_left, window_size)
smooth_angle_x_right = moving_average(angle_x_right, window_size)
smooth_angle_y_right = moving_average(angle_y_right, window_size)
smooth_angle_z_right = moving_average(angle_z_right, window_size)

# Plot angle data
plt.figure(figsize=(19, 9))

plt.subplot(3, 1, 1)
plt.plot(time[window_size-1:], smooth_angle_x_left, label='Smoothed Angle_X_Left (deg)')
plt.plot(time[window_size-1:], smooth_angle_x_right, label='Smoothed Angle_X_Right (deg)')
plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.title('Angle X vs. Time')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time[window_size-1:], smooth_angle_y_left, label='Smoothed Angle_Y_Left (deg)')
plt.plot(time[window_size-1:], smooth_angle_y_right, label='Smoothed Angle_Y_Right (deg)')
plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.title('Angle Y vs. Time')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time[window_size-1:], smooth_angle_z_left, label='Smoothed Angle_Z_Left (deg)')
plt.plot(time[window_size-1:], smooth_angle_z_right, label='Smoothed Angle_Z_Right (deg)')
plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.title('Angle Z vs. Time')
plt.legend()

plt.tight_layout()
plt.show()
```
